# Gráfica de las funciones

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 4*x + 2
x = np.linspace(-2, 6, 400)
y = f(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

## Función 2

```python
def f(x):
    return np.sin(3*x)
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
y = f(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

## Función 3

```python
def f(x):
    return (2.71828)**(2*x) - 3
x = np.linspace(-2, 2, 400)
y = f(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

## Función 4

```python
def f(x):
    return np.sqrt(x**2 - 4)

x = np.linspace(-3, 3, 400)
y = f(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

## Función 5

```python
def f(x):
    return np.cos(x + np.pi/4)

x = np.linspace(-2*np.pi, 2*np.pi, 400)
y = f(x)

plt.plot(x, y)
plt.show()
```

## Grupo 1

```python
def y1(x):
    return np.sin(x)

def y2(x):
    return np.cos(x)

x = np.linspace(-2*np.pi, 2*np.pi, 400)
y1 = y1(x)
y2 = y2(x)

# Graficar las funciones
plt.plot(x, y1, label='y1 = sin(x)')
plt.plot(x, y2, label='y2 = cos(x)')
plt.show()
```
